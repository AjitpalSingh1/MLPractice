```python
# Standardize dealership and make names
goauto_df['dealer_clean'] = goauto_df['dealer_name'].astype(str).str.strip().str.upper()
goauto_df['make_clean'] = goauto_df['make'].astype(str).str.strip().str.upper()

# Explicit dealership-to-brand mapping
dealer_make_groups = {
    'GO HONDA': {'HONDA'},
    'GO NISSAN SOUTH': {'NISSAN'},
    'JAGUAR LAND ROVER EDMONTON': {'JAGUAR', 'LAND ROVER'},
    'KENTWOOD FORD': {'FORD'},
    'NORDEN VOLKSWAGEN': {'VOLKSWAGEN'},
    'PORSCHE CENTRE EDMONTON': {'PORSCHE'},
    'TOYOTA ON THE TRAIL': {'TOYOTA'}
}

# Optional readable column for checking
goauto_df['dealer_makes'] = goauto_df['dealer_clean'].map(
    lambda d: ', '.join(sorted(dealer_make_groups.get(d, set())))
)

# Classify on-make vs off-make
def classify_service_type(row):
    dealer = row['dealer_clean']
    make = row['make_clean']
    allowed_makes = dealer_make_groups.get(dealer, set())

    if make in allowed_makes:
        return 'On-Make'
    else:
        return 'Off-Make'

goauto_df['service_type'] = goauto_df.apply(classify_service_type, axis=1)
```

```python
Q1 = goauto_df["cost"].quantile(0.25)
Q3 = goauto_df["cost"].quantile(0.75)
IQR = Q3 - Q1

lower_cost = Q1 - 1.5 * IQR
upper_cost = Q3 + 1.5 * IQR

goauto_df["cost_winsorized"] = goauto_df["cost"].clip(
    lower=lower_cost,
    upper=upper_cost)
```

```python
Q3_m = goauto_df["mileage"].quantile(0.75)
IQR_m = goauto_df["mileage"].quantile(0.75) - goauto_df["mileage"].quantile(0.25)
upper_mileage = Q3_m + 1.5 * IQR_m

goauto_df["mileage_capped"] = goauto_df["mileage"].clip(upper=upper_mileage)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.boxplot(x=goauto_df["cost"])
plt.title("Service Cost (Before Outlier Treatment)")

plt.subplot(1,2,2)
sns.boxplot(x=goauto_df["cost_winsorized"])
plt.title("Service Cost (After Winsorization)")

plt.tight_layout()
plt.show()
```

```python
goauto_df["service_date"] = pd.to_datetime(goauto_df["service_date"], errors="coerce")
goauto_df = goauto_df.dropna(subset=["service_date","cost"])

goauto_df["service_year"]  = goauto_df["service_date"].dt.year.astype(int)
goauto_df["service_month"] = goauto_df["service_date"].dt.month.astype(int)

monthly_df = (goauto_df.groupby(["dealer_name","service_year","service_month"], as_index=False).agg(
        total_revenue=("cost","sum"),
        service_count=("cost","size"),
        avg_revenue_per_service=("cost","mean")))
```

```python
monthly_df = monthly_df.sort_values(["dealer_name","service_year","service_month"])
monthly_df["revenue_lag1"] = monthly_df.groupby("dealer_name")["total_revenue"].shift(1)
monthly_df["revenue_lag3"] = monthly_df.groupby("dealer_name")["total_revenue"].shift(3)
monthly_df["revenue_roll3"] = (monthly_df.groupby("dealer_name")["total_revenue"]
                               .shift(1).rolling(3).mean())
monthly_df = monthly_df.dropna()
```

```python
monthly_df = monthly_df.dropna()
```

```python
monthly_df = pd.get_dummies(
    monthly_df,
    columns=["dealer_name"],
    drop_first=True)
```

```python
X = monthly_df.drop(columns=["total_revenue"])
y = monthly_df["total_revenue"]
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import r2_score
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

target_col = "total_revenue"

tscv = TimeSeriesSplit(n_splits=5)
r2_scores = []

for train_idx, test_idx in tscv.split(monthly_df):
    train_df = monthly_df.iloc[train_idx]
    test_df  = monthly_df.iloc[test_idx]

    X_train = train_df.drop(columns=[target_col])
    y_train = train_df[target_col]
    X_test  = test_df.drop(columns=[target_col])
    y_test  = test_df[target_col]
```

```python
def evaluate(y_true, y_pred, name):

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)

    print(f"Model: {name}")
    print(f"MAE : {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R2  : {r2:.3f}")
    print("-"*30)
```

```python
# Baseline Model (Naive Forecast)

baseline_pred = X_test["revenue_lag1"]

evaluate(y_test, baseline_pred, "Naive Baseline")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train,y_train)

lr_pred = lr_model.predict(X_test)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=200)

rf_model.fit(X_train,y_train)

rf_pred = rf_model.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

gbr_default = GradientBoostingRegressor(random_state=42)

gbr_default.fit(X_train, y_train)

gbr_default_pred = gbr_default.predict(X_test)

mae = mean_absolute_error(y_test, gbr_default_pred)
rmse = np.sqrt(mean_squared_error(y_test, gbr_default_pred))
r2 = r2_score(y_test, gbr_default_pred)

print("Default Gradient Boosting")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
gbr = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,)

gbr.fit(X_train, y_train)
gbr_pred = gbr.predict(X_test)

mae = mean_absolute_error(y_test, gbr_pred)
rmse = np.sqrt(mean_squared_error(y_test, gbr_pred))
r2 = r2_score(y_test, gbr_pred)

print("Gradient Boosting")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```
